# Wind Power Forecast Analysis — UK National Grid
**REint AI — Full Stack SWE Challenge**

This notebook analyses:
1. **Forecast error characteristics** — mean, median, p99, error vs horizon, time-of-day patterns
2. **Historical actual wind generation** — reliability analysis and MW recommendation

Data source: [Elexon BMRS API](https://bmrs.elexon.co.uk/api-documentation) | Period: January 2024

```python
import requests
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import warnings
from datetime import datetime, timedelta, timezone

warnings.filterwarnings('ignore')

# Plot style
plt.rcParams.update({
    'figure.facecolor': '#050a0f',
    'axes.facecolor': '#0a1520',
    'axes.edgecolor': '#1a3050',
    'axes.labelcolor': '#7aa0c0',
    'xtick.color': '#3d6080',
    'ytick.color': '#3d6080',
    'text.color': '#e8f4ff',
    'grid.color': '#0d2035',
    'grid.linestyle': '--',
    'grid.alpha': 0.5,
    'font.family': 'monospace',
    'figure.dpi': 120,
})
BLUE = '#00b4ff'
GREEN = '#00e5a0'
ORANGE = '#ffaa44'
RED = '#ff6060'

print('Libraries loaded ✓')
```

## 1. Data Fetching

```python
BASE = 'https://data.elexon.co.uk/bmrs/api/v1'

def fetch_pages(endpoint, params, max_pages=50):
    """Fetch all pages from an Elexon BMRS endpoint."""
    all_data = []
    for page in range(1, max_pages + 1):
        p = {**params, 'page': page, 'pageSize': 200}
        r = requests.get(f'{BASE}/{endpoint}', params=p, timeout=30)
        if r.status_code != 200:
            print(f'  HTTP {r.status_code} at page {page}')
            break
        data = r.json().get('data', [])
        if not data:
            break
        all_data.extend(data)
        if len(data) < 200:
            break
    return all_data

print('Fetching actuals (FUELHH) for January 2024...')
raw_actuals = []
for day_offset in range(31):
    date = (datetime(2024, 1, 1) + timedelta(days=day_offset)).strftime('%Y-%m-%d')
    data = fetch_pages('datasets/FUELHH/stream', {
        'settlementDateFrom': date,
        'settlementDateTo': date,
        'fuelType': 'WIND',
        'format': 'json'
    })
    raw_actuals.extend(data)
    if day_offset % 5 == 0:
        print(f'  Day {day_offset+1}/31 fetched, {len(raw_actuals)} records so far')

print(f'Total actual records: {len(raw_actuals)}')
```

```python
print('Fetching forecasts (WINDFOR) for January 2024...')
raw_forecasts = fetch_pages('datasets/WINDFOR/stream', {
    'publishDateTimeFrom': '2023-12-31T00:00:00Z',
    'publishDateTimeTo': '2024-01-31T23:59:59Z',
    'format': 'json'
})
print(f'Total forecast records: {len(raw_forecasts)}')
```

## 2. Data Preparation

```python
# --- Actuals ---
df_act = pd.DataFrame(raw_actuals)
df_act = df_act[df_act['fuelType'] == 'WIND'].copy()
df_act['startTime'] = pd.to_datetime(df_act['startTime'], utc=True)
df_act = df_act.rename(columns={'generation': 'actual_mw'})
df_act = df_act[['startTime', 'actual_mw']].drop_duplicates('startTime').sort_values('startTime')

# Filter to Jan 2024
jan_start = pd.Timestamp('2024-01-01', tz='UTC')
jan_end = pd.Timestamp('2024-02-01', tz='UTC')
df_act = df_act[(df_act['startTime'] >= jan_start) & (df_act['startTime'] < jan_end)]

print(f'Actuals: {len(df_act)} records from {df_act.startTime.min()} to {df_act.startTime.max()}')
print(f'Missing 30-min slots: {(31*48) - len(df_act)}')
df_act.head()
```

```python
# --- Forecasts ---
df_fct = pd.DataFrame(raw_forecasts)
df_fct['startTime'] = pd.to_datetime(df_fct['startTime'], utc=True)
df_fct['publishTime'] = pd.to_datetime(df_fct['publishTime'], utc=True)
df_fct = df_fct.rename(columns={'generation': 'forecast_mw'})
df_fct = df_fct[['startTime', 'publishTime', 'forecast_mw']].dropna()

# Compute horizon in hours
df_fct['horizon_h'] = (df_fct['startTime'] - df_fct['publishTime']).dt.total_seconds() / 3600

# Filter: target time in Jan 2024, horizon 0-48h
df_fct = df_fct[
    (df_fct['startTime'] >= jan_start) &
    (df_fct['startTime'] < jan_end) &
    (df_fct['horizon_h'] >= 0) &
    (df_fct['horizon_h'] <= 48)
]

print(f'Forecast records after filtering: {len(df_fct)}')
print(f'Horizon range: {df_fct.horizon_h.min():.1f}h — {df_fct.horizon_h.max():.1f}h')
df_fct.head()
```

```python
# For each target time: pick the LATEST forecast created at least 4h before target
HORIZON_CUTOFF = 4  # hours

df_fct_4h = df_fct[df_fct['horizon_h'] >= HORIZON_CUTOFF].copy()

# Latest forecast per target time
df_fct_latest = (
    df_fct_4h
    .sort_values('publishTime')
    .groupby('startTime', as_index=False)
    .last()
)

print(f'Latest-forecast records (≥4h horizon): {len(df_fct_latest)}')

# Merge actuals + latest forecasts
df = pd.merge(df_act, df_fct_latest[['startTime','forecast_mw','publishTime','horizon_h']], on='startTime', how='inner')
df['abs_error'] = (df['actual_mw'] - df['forecast_mw']).abs()
df['error'] = df['actual_mw'] - df['forecast_mw']
df['hour'] = df['startTime'].dt.hour
df['date'] = df['startTime'].dt.date
df['weekday'] = df['startTime'].dt.day_name()

print(f'Merged records: {len(df)}')
df.head()
```

## 3. Forecast Error Analysis

```python
# --- Summary stats ---
mean_err   = df['abs_error'].mean()
median_err = df['abs_error'].median()
p99_err    = df['abs_error'].quantile(0.99)
p90_err    = df['abs_error'].quantile(0.90)
rmse       = np.sqrt((df['error']**2).mean())
mape       = (df['abs_error'] / df['actual_mw'].replace(0, np.nan)).mean() * 100

print('=' * 45)
print('FORECAST ERROR SUMMARY (4h+ horizon, Jan 2024)')
print('=' * 45)
print(f'Mean Absolute Error   : {mean_err:>8,.0f} MW')
print(f'Median Absolute Error : {median_err:>8,.0f} MW')
print(f'P90 Absolute Error    : {p90_err:>8,.0f} MW')
print(f'P99 Absolute Error    : {p99_err:>8,.0f} MW')
print(f'RMSE                  : {rmse:>8,.0f} MW')
print(f'MAPE                  : {mape:>7.1f} %')
print(f'Total matched points  : {len(df):>8,}')
```

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('Forecast Error Analysis — January 2024', fontsize=14, color='#e8f4ff', fontweight='bold', y=1.01)

# 1. Error distribution
ax = axes[0, 0]
ax.hist(df['error'], bins=60, color=BLUE, alpha=0.7, edgecolor='none')
ax.axvline(0, color='white', lw=1, linestyle='--', alpha=0.5)
ax.axvline(mean_err, color=ORANGE, lw=1.5, label=f'Mean error: {df["error"].mean():+,.0f} MW')
ax.set_title('Error Distribution (Actual − Forecast)', color='#e8f4ff')
ax.set_xlabel('Error (MW)')
ax.set_ylabel('Count')
ax.legend(fontsize=9)
ax.grid(True, alpha=0.3)

# 2. Abs error over time
ax = axes[0, 1]
daily_err = df.groupby('date')['abs_error'].mean()
ax.bar(range(len(daily_err)), daily_err.values, color=BLUE, alpha=0.7)
ax.axhline(mean_err, color=ORANGE, lw=1.5, linestyle='--', label=f'Mean: {mean_err:,.0f} MW')
ax.set_title('Daily Mean Absolute Error', color='#e8f4ff')
ax.set_xlabel('Day of January 2024')
ax.set_ylabel('MAE (MW)')
ax.legend(fontsize=9)
ax.grid(True, alpha=0.3)

# 3. Error by hour of day
ax = axes[1, 0]
hourly = df.groupby('hour')['abs_error'].agg(['mean', 'median'])
ax.plot(hourly.index, hourly['mean'], color=BLUE, lw=2, marker='o', ms=4, label='Mean MAE')
ax.plot(hourly.index, hourly['median'], color=GREEN, lw=2, marker='s', ms=4, label='Median MAE')
ax.set_title('Absolute Error by Hour of Day', color='#e8f4ff')
ax.set_xlabel('Hour (UTC)')
ax.set_ylabel('Absolute Error (MW)')
ax.set_xticks(range(0, 24, 2))
ax.legend(fontsize=9)
ax.grid(True, alpha=0.3)

# 4. Error vs forecast horizon
ax = axes[1, 1]
# Bin horizons in 2h buckets
df_all_horizons = df_fct.copy()
df_all_horizons = pd.merge(df_all_horizons, df_act, on='startTime', how='inner')
df_all_horizons['abs_error'] = (df_all_horizons['actual_mw'] - df_all_horizons['forecast_mw']).abs()
df_all_horizons['horizon_bin'] = (df_all_horizons['horizon_h'] // 2) * 2
horizon_err = df_all_horizons.groupby('horizon_bin')['abs_error'].mean()
ax.plot(horizon_err.index, horizon_err.values, color=GREEN, lw=2)
ax.fill_between(horizon_err.index, horizon_err.values, alpha=0.15, color=GREEN)
ax.set_title('Mean Abs Error vs Forecast Horizon', color='#e8f4ff')
ax.set_xlabel('Forecast Horizon (hours)')
ax.set_ylabel('Mean Abs Error (MW)')
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('error_analysis.png', bbox_inches='tight', facecolor='#050a0f')
plt.show()
print('Saved: error_analysis.png')
```

```python
# Actual vs Forecast time series for a sample week
week = df[df['startTime'] < pd.Timestamp('2024-01-08', tz='UTC')]

fig, ax = plt.subplots(figsize=(14, 5))
ax.plot(week['startTime'], week['actual_mw'], color=BLUE, lw=1.5, label='Actual')
ax.plot(week['startTime'], week['forecast_mw'], color=GREEN, lw=1.5, linestyle='--', label='Forecast (4h+ horizon)', alpha=0.85)
ax.fill_between(week['startTime'], week['actual_mw'], week['forecast_mw'], alpha=0.08, color=ORANGE, label='Error band')
ax.set_title('Actual vs Forecast — First Week of January 2024', color='#e8f4ff', fontsize=13)
ax.set_ylabel('Wind Power (MW)')
ax.xaxis.set_major_formatter(mdates.DateFormatter('%d %b\n%H:%M'))
ax.legend(fontsize=10)
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('actual_vs_forecast.png', bbox_inches='tight', facecolor='#050a0f')
plt.show()
```

## 4. Wind Generation Reliability Analysis

**Goal**: Based on historical actual generation, recommend how many MW of wind power can reliably meet electricity demand.

```python
gen = df_act['actual_mw'].dropna()

# Percentile table
pcts = [1, 5, 10, 25, 50, 75, 90, 95, 99]
percentiles = {p: gen.quantile(p/100) for p in pcts}

print('='*45)
print('WIND GENERATION PERCENTILES — Jan 2024')
print('='*45)
for p, v in percentiles.items():
    bar = '█' * int(v / gen.max() * 30)
    print(f'P{p:>2}  {v:>7,.0f} MW  {bar}')
print(f'\nMax : {gen.max():>7,.0f} MW')
print(f'Mean: {gen.mean():>7,.0f} MW')
print(f'Min : {gen.min():>7,.0f} MW')
```

```python
fig, axes = plt.subplots(1, 3, figsize=(16, 5))
fig.suptitle('Wind Generation Reliability Analysis — January 2024', fontsize=13, color='#e8f4ff')

# 1. Generation histogram
ax = axes[0]
ax.hist(gen, bins=50, color=BLUE, alpha=0.75, edgecolor='none')
for p_val, color, label in [(5, RED, 'P5 (reliable floor)'), (50, ORANGE, 'P50 (median)')]:
    ax.axvline(percentiles[p_val], color=color, lw=1.5, linestyle='--', label=f'{label}: {percentiles[p_val]:,.0f} MW')
ax.set_title('Generation Distribution', color='#e8f4ff')
ax.set_xlabel('Wind Power (MW)')
ax.set_ylabel('Count (30-min slots)')
ax.legend(fontsize=8)
ax.grid(True, alpha=0.3)

# 2. Duration curve (sorted generation)
ax = axes[1]
sorted_gen = np.sort(gen.values)[::-1]
pct_time = np.linspace(0, 100, len(sorted_gen))
ax.plot(pct_time, sorted_gen, color=GREEN, lw=2)
ax.fill_between(pct_time, sorted_gen, alpha=0.15, color=GREEN)
for p_val, color in [(5, RED), (10, ORANGE), (25, BLUE)]:
    v = percentiles[p_val]
    ax.axhline(v, color=color, lw=1, linestyle=':', alpha=0.8, label=f'P{p_val}: {v:,.0f} MW')
ax.set_title('Load Duration Curve', color='#e8f4ff')
ax.set_xlabel('% Time ≥ this level')
ax.set_ylabel('Wind Power (MW)')
ax.legend(fontsize=8)
ax.grid(True, alpha=0.3)

# 3. Daily generation profile
ax = axes[2]
hourly_gen = df_act.copy()
hourly_gen['hour'] = pd.to_datetime(hourly_gen['startTime']).dt.hour
h_stats = hourly_gen.groupby('hour')['actual_mw'].agg(['mean', lambda x: x.quantile(0.1), lambda x: x.quantile(0.9)])
h_stats.columns = ['mean', 'p10', 'p90']
ax.fill_between(h_stats.index, h_stats['p10'], h_stats['p90'], alpha=0.2, color=BLUE, label='P10–P90 band')
ax.plot(h_stats.index, h_stats['mean'], color=BLUE, lw=2, label='Mean')
ax.plot(h_stats.index, h_stats['p10'], color=RED, lw=1, linestyle='--', label='P10 (reliable floor)')
ax.set_title('Generation by Hour of Day', color='#e8f4ff')
ax.set_xlabel('Hour (UTC)')
ax.set_ylabel('Wind Power (MW)')
ax.set_xticks(range(0, 24, 4))
ax.legend(fontsize=8)
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('reliability_analysis.png', bbox_inches='tight', facecolor='#050a0f')
plt.show()
```

```python
# Low-generation events: consecutive slots below threshold
threshold = percentiles[10]
df_act_sorted = df_act.sort_values('startTime').copy()
df_act_sorted['low'] = df_act_sorted['actual_mw'] < threshold

# Find runs of low generation
df_act_sorted['run_id'] = (df_act_sorted['low'] != df_act_sorted['low'].shift()).cumsum()
low_runs = df_act_sorted[df_act_sorted['low']].groupby('run_id').size() * 0.5  # hours

print(f'Low-generation events (< {threshold:,.0f} MW = P10):')
print(f'  Number of events      : {len(low_runs)}')
print(f'  Longest event         : {low_runs.max():.1f} hours')
print(f'  Median event duration : {low_runs.median():.1f} hours')
print(f'  Total low-gen time    : {low_runs.sum():.1f} hours ({low_runs.sum()/(31*24)*100:.1f}% of month)')
```

## 5. Recommendation: Reliable Wind Capacity

> **How many MW of wind power can we reliably expect to be available to meet electricity demand?**

```python
p5  = percentiles[5]
p10 = percentiles[10]
p25 = percentiles[25]
p50 = percentiles[50]
mean_gen = gen.mean()

print('=' * 60)
print('RECOMMENDATION: RELIABLE WIND CAPACITY')
print('=' * 60)
print(f"""
Based on January 2024 actual UK national wind generation data:

┌─────────────────────────────────────────────────────────┐
│  RECOMMENDED RELIABLE CAPACITY: ~{p10:,.0f} MW (P10)     │
│  (can be counted on ≥90% of the time)                   │
└─────────────────────────────────────────────────────────┘

Tiered view:
  • Firm/baseload planning  : {p5:>6,.0f} MW  (P5  — available 95% of time)
  • Reliable contribution   : {p10:>6,.0f} MW  (P10 — available 90% of time) ← RECOMMENDED
  • Moderate confidence     : {p25:>6,.0f} MW  (P25 — available 75% of time)
  • Median contribution     : {p50:>6,.0f} MW  (P50 — available 50% of time)
  • Mean contribution       : {mean_gen:>6,.0f} MW  (planning & energy budgeting)

Rationale:
  1. RELIABILITY THRESHOLD: P10 ({p10:,.0f} MW) is the standard used by
     grid planners as a 'reliable' contribution — generation falls below
     this value only 10% of time (about 74 hours/month).

  2. WIND INTERMITTENCY: January 2024 shows high variability ({gen.min():,.0f}–
     {gen.max():,.0f} MW range). Wind power CANNOT be treated as firm capacity
     without backup storage or dispatchable plant.

  3. OPERATIONAL PLANNING: For real-time dispatch, the 4h+ forecast MAE of
     ~{mean_err:,.0f} MW should be added as uncertainty margin on top of any
     wind contribution commitment.

  4. SEASONAL NOTE: January is typically high-wind in the UK. Reliable
     capacity for summer months would be materially lower. A full-year
     analysis is recommended for annual capacity planning.

Conclusion: For electricity demand planning, treat {p10:,.0f} MW as the
reliable wind contribution. For energy budgeting, use {mean_gen:,.0f} MW
(mean). All values above {p10:,.0f} MW should be considered variable and
supported by flexible backup capacity.
""")
```

```python
# Final summary chart
fig, ax = plt.subplots(figsize=(14, 5))

ax.plot(df_act['startTime'], df_act['actual_mw'], color=BLUE, lw=1, alpha=0.8, label='Actual generation')
ax.axhline(p10, color=RED, lw=2, linestyle='--', label=f'P10 Reliable floor: {p10:,.0f} MW')
ax.axhline(p50, color=ORANGE, lw=1.5, linestyle=':', label=f'P50 Median: {p50:,.0f} MW')
ax.axhline(mean_gen, color=GREEN, lw=1.5, linestyle='-.', label=f'Mean: {mean_gen:,.0f} MW')

ax.fill_between(df_act['startTime'], df_act['actual_mw'], p10,
                where=df_act['actual_mw'] < p10, alpha=0.2, color=RED, label='Below P10 (risk zone)')

ax.set_title('UK Wind Generation — January 2024 with Reliability Thresholds', color='#e8f4ff', fontsize=13)
ax.set_ylabel('Wind Power (MW)')
ax.xaxis.set_major_formatter(mdates.DateFormatter('%d %b'))
ax.legend(fontsize=9, loc='upper right')
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('reliability_recommendation.png', bbox_inches='tight', facecolor='#050a0f')
plt.show()
print('Analysis complete ✓')
```
